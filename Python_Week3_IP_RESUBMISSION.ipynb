### Telcom Datasets

```python
# To confirm column names in all 3 datasets match
columns = ['PRODUCT', 'VALUE', 'DATE_TIME', 'CELL_ON_SITE', 'DW_A_NUMBER', 'DW_B_NUMBER', 'COUNTRY_A', 'COUNTRY_B','CELL_ID', 'SITE_ID']
datasets = [df1, df2, df3]

for data in datasets:
  data.columns = columns
  data.columns = map(str.lower, data.columns)
```

```python
# Concat the datasets
telcom = pd.concat([df1, df2, df3])
telcom.reset_index(drop = True, inplace = True)
telcom.head()
```

```python
telcom.info()
```

## Merging Data

```python
# Checking if similar columns exist in our now two datasets
def similar_columns(data1, data2, col1, col2):
  data1[col1] = data1[col1].str.strip()
  data2[col2] = data2[col2].str.strip()

  fisrt_list = data1[col1].to_list()
  second_list = data2[col2].to_list()

  check = any(item in fisrt_list for item in second_list)

  return check

similar_columns(df_cells, telcom, 'cell_id', 'cell_id')


```

```python
similar_columns(df_cells, telcom, 'site_code', 'site_id')
```

```python
# Merge telcom dataset with cells geo data set on site_code and site_id
df = pd.merge(telcom, df_cells, how = 'inner', left_on='site_id', right_on= 'site_code')
df.head()
```

```python
# Confirming dates
def confirm_dates(data,col):
  data[col] = data[col].apply(lambda x:
                              dt.datetime.strptime(x, '%Y-%m-%d %H:%M:%S.%f'))
  data['DAY'] = pd.to_datetime(data[col]).dt.day
  output = data['DAY'].value_counts()
  data.drop('DAY',axis = 1, inplace = True)

  return output

confirm_dates(df, 'date_time')
```

```python
# Drop all irrelevant columns for our analysis
def dropping_columns(data,cols):
  data.drop(cols, axis = 1, inplace = True)

  return data

columns = ['longitude', 'latitude', 'cell_id_x', 'cell_id_y', 'cell_on_site', 'country_a', 'country_b',]
dropping_columns(df, columns)
```

- columns longitude and latitude are used to provide location which is already provided by localisation
- cell_id not required in our analysis
- columns: country_a, country_b and cell_on site are described as unnecessary
