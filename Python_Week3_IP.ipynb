## Importing Libraries to be used

```python
# Import pandas library
# as well as numpy library
import pandas as pd
import numpy as np
```

## Loading Our Data Sets

```python
# Load cells_geo csv file
#
df_cells = pd.read_csv('cells_geo.csv', ';', header = 1, names = ['City', 'Status', 'Localisation', 'Geographical_Zone', 'Zone_Name', 'Longitude', 'Latitude', 'Region', 'Area', 'Cell_ID', 'Site_ID'])
df_cells
```

```python
# Load the Telcom Dataset1
# As it is a csv file, load it into the database as shown
#

df1= pd.read_csv ('Telcom_dataset.csv')
df1
```

```python
# Load Telcom Dataset2
#
df2 = pd.read_csv('Telcom_dataset2.csv')
df2
```

```python
# Load Telcom Dataset3
#
df3 = pd.read_csv('Telcom_dataset3.csv')
df3
```

## Which ones were the most used city for the three days?

```python
# To diplay the most used cities for the three days
#
df_cells.groupby('City').count().nlargest(5,'Status')
```

## Which Product was most used for the 3 days?

```python
# #Combine the CDR datasets
df_cdr = pd.concat([df1,df2,df3])

#To find out the product used the most for the 3 days
#
df_cdr.groupby('PRODUCT').count().nlargest(5,'VALUE')
```
